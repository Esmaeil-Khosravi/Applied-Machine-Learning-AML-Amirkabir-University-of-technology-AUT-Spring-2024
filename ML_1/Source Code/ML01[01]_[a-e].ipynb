# Handeling outliers:

#### A function for detecting the range of outlier values baded on the  [Q1-1.5*IQR, Q3+1.5*IQR] method has been coded:

#### This function help us determine the range of outlier values of every attribute of our dataset.

```python
import numpy as np
 
def detect_outlier(data):
    q1, q3 = np.percentile(sorted(data), [25, 75])
    iqr = q3 - q1
    lower_bound = q1 - (1.5 * iqr)
    upper_bound = q3 + (1.5 * iqr)
    outliers = [x for x in data if x <= lower_bound or x >= upper_bound]
 
    return [lower_bound, upper_bound]
```

```python
detect_outlier(data['age'])
```

#### the above range shows the outlier range of age att.
